```python
import speech_recognition as sr
import pyttsx3
import pywhatkit as pwk
import datetime
import os
import cv2
import webbrowser
import sys
import PyPDF2
import xlsxwriter


engine=pyttsx3.init()  

def talk(text):   
    engine.say(text)
    print(text)
    engine.runAndWait()
    
def alexa_command():
    r=sr.Recognizer()
    with sr.Microphone() as source:
        print("listener...")
        r.pause_threshold = 2
        voice=r.listen(source,timeout=1,phrase_time_limit=7)

    try:
        print("recognizing...")
        command=r.recognize_google(voice, language="en-in")
        print(f"user said : {command}")
        talk(command)
        command=command.lower()
        if "alexa" in command:
            command=command.replace("alexa","")
            print(command)
    except Exception as e:
        talk("say that again please,babu")
        return "none"
    return command

def wish():
    time=int(datetime.datetime.now().hour)
    if time>=0 and time<=12:
        talk("good morning my babu")
    elif time>12 and time<=18:
        talk("good afternoon my babu")
    else:
        talk("good night my babu")
        
        
def pdf_reader():
    book=open("E:\\DOCUMENTS\\resume1.pdf",'rb')
    pdfReader=PyPDF2.PdfFileReader(book)
    pages=pdfReader.numPages
    talk(f"babu in this pdf there are {pages}")
    talk("which you have to read babu")
    pg=int(input("enter page number:"))
    page=pdfReader.getPage(pg)
    text=page.extractText()
    talk(text)
    
def wrt_voice():
    f = open("E:\\New folder\\ttt.txt", "a")
    talk("please tell me babu which is write in notepad")
    st=alexa_command().lower()
    f.write(str(st))
    f.close()
    
def wte_excel():
    workbook = xlsxwriter.Workbook('E:\\New folder\\ppt.xlsx',"r")
    worksheet = workbook.add_worksheet()

    talk("please tell me 1st heading")
    st=alexa_command().lower()
    worksheet.write('A1',str(st))
    
    talk('what is your : '+st)
    st=alexa_command().lower()
    worksheet.write('A2',str(st))

    talk("please tell me 2nd heading")
    st=alexa_command().lower()
    worksheet.write('B1',str(st))
    
    talk('what is your : '+st)
    st=alexa_command().lower()
    worksheet.write('B2',str(st))

    talk("do you have 3rd heading")
    pd=alexa_command().lower()
    if pd=="yes" or pd=="haa":
        talk("please 3rd heading")
        st=alexa_command().lower()
        worksheet.write('C1',str(st))
        
        talk('what is your : '+st)
        st=alexa_command().lower()
        worksheet.write('C2',str(st))
    if pd=="no" or pd=="nhi":
        pass


    talk("do you have 4rt heading")
    pd=alexa_command().lower()
    if pd=="yes" or pd=="haa":
        talk("please 4rt heading")
        st=alexa_command().lower()
        worksheet.write('D1',str(st))
        
        talk("please tell me D2 heading")
        st=alexa_command().lower()
        worksheet.write('D2',str(st)) 
    if pd=="no" or pd=="nhi":
        pass


   

   


   

   
    
    talk("your data successful added")

    workbook.close()



if __name__ == "__main__":
    wish()
    talk("i am your assistance. tell me may i help you babu")
    #alexa_command()
    command=alexa_command()
    

    if "youtube" in command:
        song=command.replace("youtube","")
        talk("your youtube song is "+song)
        pwk.playonyt(song)

    elif "time" in command:
        time=datetime.datetime.now().strftime("%H:%M")
        t=command.replace("what is time","")
        print("current time is:",time)
        talk("abhi ka time hai ki"+time)

    elif "play music" in command:
        music="E:\\NIRDESH KUMAR C\\SONGS"
        songs=os.listdir(music)
        print(songs)
        os.startfile(os.path.join(music,songs[0]))    

    elif "open cmp" in command:
        talk("babu, what should i search on google")
        cm=alexa_command().lower()
        if cm=="yes":
            os.system("start cmd")
        else:
            pass


    elif "open camera" in command:
   
        cap=cv2.VideoCapture(0)

        ret,img=cap.read()
        cv2.imshow("webcam",img)
        k=cv2.waitKey(50)


        cap.release()
        cv2.destroyAllWindows()

    elif "open google" in command:
        talk("babu, what should i search on google")
        cm=alexa_command().lower()
        webbrowser.open(f"{cm}")

    elif "send message" in command:
        talk("babu, what want i send message on whatsapp")
        odr=alexa_command().lower()
        pwk.sendwhatmsg("919837537303",{odr},5,30)
        
        

    elif "read pdf" in command or "read" in command:
        pdf_reader()
        

    elif "open excel" in command:
        talk("are you write in excel")
        wte_excel()

    elif "open notepad" in command:
        wrt_voice()
     
    
```
